## Principal Component Analysis (PCA)

http://scikit-learn.org/stable/modules/generated/sklearn.decomposition.PCA.html

```python
# TODO 
# X is all the questions but c we don't know, we want to find similarities
X = df
#c = no c because data is unsupervised - we want to find the underlie features of the data
```

```python
pca = decomposition.PCA().fit(X)
```

```python
pca.n_components_
```

### Explained Variance

```python
normalized_eigenvalues = pca.explained_variance_ratio_
```

```python
normalized_eigenvalues
# this is a normalized variance ratio- there are more weigh on the first or second values - 0.34 and 0.179
# 1/3 or 34% of 62 questions , second questions 17%...it order from the highest percentage to the lowest but it is
# indepedent from the order of the questions
```

```python
normalized_eigenvalues.sum()
```

### Scree plot

```python
plt.plot(normalized_eigenvalues)

plt.scatter(x = range(len(normalized_eigenvalues)),
    y = normalized_eigenvalues,
    marker = '+',
    s = 100)

plt.xlabel('Explained Variance')
plt.ylabel('Principal Component')
plt.xlim(-1, len(normalized_eigenvalues))
plt.ylim(-.01, max(normalized_eigenvalues) + .01)
```

```python
# the first four questions with highest weight are shown in the graphic
```

```python
cumulated_normalized_eigenvalues = np.cumsum(normalized_eigenvalues)
```

### `n_components = 3`

```python
#just take the first three since they are more important
```

```python
X = df

# TODO
pca = decomposition.PCA(n_components = 3).fit(X)

X = pca.transform(X)

# TODO
X = pd.DataFrame({'PC0': X.T[0], 'PC1': X.T[1], 'PC2': X.T[2]})
```

```python
X
```

```python
X.shape
# all information compressed into 3 axis
```

```python
eigenvector_0 = pca.components_[0]
eigenvector_1 = pca.components_[1]
eigenvector_2 = pca.components_[2]
```

```python
# look at the plus and minus of the questions
def interpret(v):
    v = pd.Series(v).sort_values(ascending = False)
   
    for question in questions[v.head(5).index]:
        print '+ {}'. format(question)
       
        print
       
    for question in questions[v.tail(5).index]:
        print '- {}'. format(question)
```
